```python
import numpy as np
import pandas as pd
import json
import sys
import os
import kashgari
from kashgari.embeddings import BERTEmbedding
from kashgari.tasks.labeling import BiLSTM_CRF_Model
from tqdm import tqdm
import math
CUT_LENGTH = 400
```

```python
tag = {}
tag['疾病和诊断']=['S-DEA','E-DEA']
tag['影像检查']=['S-PIC','E-PIC']
tag['实验室检验']=['S-LAB','E-LAB']
tag['药物']=['S-MED','E-MED']
tag['手术']=['S-OPR','E-OPR']
tag['解剖部位']=['S-POS','E-POS']
tag_hash={}
tag_hash['DEA']='疾病和诊断'
tag_hash['PIC']='影像检查' 
tag_hash['LAB']='实验室检验'
tag_hash['MED']='药物'
tag_hash['OPR']='手术'
tag_hash['POS']='解剖部位'
```

```python
class Logger(object):
    def __init__(self, filename="Default.log"):
        self.terminal = sys.stdout
        self.log = open(filename, "w",encoding='utf-8')
    def write(self, message):
        self.terminal.write(message)
        self.log.write(message)
        self.log.flush()
    def flush(self):
        pass
sys.stdout = Logger('output.txt')
```

```python
def read_train_data():
    x_train = []
    y_train = []
    fp = open('train.json','r',encoding='utf-8-sig')
    datas = fp.readlines()
    fp.close()
    for data in datas:
        cut_data = []
        label = []
        a = json.loads(data)
        originaltext = a['originalText']
        entities = a['entities']
        i = 0
        while i < len(originaltext):
            cut_data.append(originaltext[i])
            label.append('0')
            i = i + 1
        j = 0
        while j < len(entities):
            flag = 0
            for k in range(entities[j]['start_pos'],entities[j]['end_pos']):
                if flag == 0:
                    label[k] = tag[entities[j]['label_type']][0]
                    flag = 1
                else:
                    label[k] = tag[entities[j]['label_type']][1]
            j = j + 1
        #print(type(label))
        x_train.append(cut_data)
        y_train.append(label)
    return x_train,y_train
```

```python
def print_data(index, res, x_test, y_test):
    flag = False
    i = 0
    while i< len(y_test):
        if y_test[i] != '0' and y_test[i] != '[PAD]' and flag == False:
            entity_category = y_test[i]
            start = i
            flag = True
            while y_test[i][-3:] == entity_category[-3:] and i < len(y_test)-1:
                i += 1
            end = i
            flag = False
            print(tag_hash[entity_category[-3:]])
            print(x_test[start:end])
            res=res.append(pd.DataFrame({'textId':[int(index)],'label_type':[tag_hash[entity_category[-3:]]],'start_pos':[start],'end_pos':[end]}), ignore_index = True)
        i += 1
    return res
```

```python
def data_segmentation(x_train,y_train):
    x_new_train = []
    y_new_train = []
    i = 0
    while i < len(x_train):   
        if len(x_train[i]) <=  CUT_LENGTH:  #综合考虑，bert每次处理句子长度设为400
            x_new_train.append(x_train[i])
            y_new_train.append(y_train[i])
        else:
            length = len(x_train[i])
            time = math.ceil(length/ CUT_LENGTH)
            limit = CUT_LENGTH - (length% CUT_LENGTH) + 1
            j = 0
            k,start,end = 0,0,CUT_LENGTH
            while j < time - 1:
                if y_train[i][end - 1] != '0':  #防止出现切分关键分词的情况
                    while y_train[i][end - 1] != '0':
                        end = end - 1
                        k = k + 1
                    x_new_train.append(x_train[i][start:end - 1])
                    y_new_train.append(y_train[i][start:end - 1])
                else:
                    x_new_train.append(x_train[i][start:start + CUT_LENGTH - 1])
                    y_new_train.append(y_train[i][start:start + CUT_LENGTH - 1])
                if k >= limit:
                    time += 1
                    k = 0
                start = end
                end = start + CUT_LENGTH
                j += 1
            x_new_train.append(x_train[i][start:])
            y_new_train.append(y_train[i][start:])
        i += 1
    return x_new_train,y_new_train
```

```python
def data_segmentation2(x_train,y_train):
    x_new_train = []
    y_new_train = []
    i = 0
    while i < len(x_train):   
        if len(x_train[i]) <=  CUT_LENGTH:  #综合考虑，bert每次处理句子长度设为400
            x_new_train.append(x_train[i])
            y_new_train.append(y_train[i])
        else:
            #print(i)
            length = len(x_train[i])
            time = math.ceil(length/ CUT_LENGTH)
            limit = CUT_LENGTH - (length% CUT_LENGTH) + 1
            j = 0
            while j < time - 1:
                if y_train[i][(j+1)*CUT_LENGTH-1] != '0':  #防止出现切分关键分词的情况
                    temp = (j+1)*CUT_LENGTH - 1
                    k = 0
                    while y_train[i][temp] != '0':
                        temp = temp - 1
                        k = k + 1
                    x_new_train.append(x_train[i][j* CUT_LENGTH:temp])
                    y_new_train.append(y_train[i][j* CUT_LENGTH:temp])
                    for t in range(k):
                        x_new_train.append('0')
                        y_new_train.append('0')
                else:    
                    x_new_train.append(x_train[i][j* CUT_LENGTH:(j+1)* CUT_LENGTH -1])
                    y_new_train.append(y_train[i][j* CUT_LENGTH:(j+1)* CUT_LENGTH -1])
                j += 1
            x_new_train.append(x_train[i][j* CUT_LENGTH:])
            y_new_train.append(y_train[i][j* CUT_LENGTH:])
        i += 1
    return x_new_train,y_new_train
```

```python
x_train,y_train = read_train_data()
print(len(x_train),len(y_train))
x_train,y_train = data_segmentation(x_train,y_train)
print(len(x_train),len(y_train))
```

```python
bert_embed = BERTEmbedding('chinese_L-12_H-768_A-12', task = kashgari.LABELING, sequence_length = CUT_LENGTH)
model = BiLSTM_CRF_Model(bert_embed)
model.fit(x_train, y_train, epochs = 20, batch_size = 32)
model.save('bert_2019')
```

```python
def read_test_data():
    x_test = []
    fp = open('test.json','r',encoding='utf-8-sig')
    datas = fp.readlines()
    fp.close()
    for data in datas:
        cut_data = []
        a = json.loads(data)
        originaltext = a['originalText']
        i = 0
        while i < len(originaltext):
            cut_data.append(originaltext[i])
            i = i + 1
        x_test.append(cut_data)
    return x_test
```

```python
x_test = read_test_data()
y_test = model.predict(x_test)
```

```python
res = pd.DataFrame(columns = ('textId','label_type','start_pos','end_pos'))
for i in (range(600)):
    res = print_data(i,res,x_test[i],y_test[i])
```
